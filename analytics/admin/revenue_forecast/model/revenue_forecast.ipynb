# Admin Revenue Forecast
Multiple Linear Regression — Predict platform-wide daily revenue (7 days).

Run cells top-to-bottom: **Setup → Generate Data → Train → Evaluate → Test**

## 1. Setup
Resolve paths for `data/`, `output/`, `report/` directories.

```python
from pathlib import Path

try:
    NOTEBOOK_DIR = Path(__vsc_ipynb_file__).resolve().parent
except NameError:
    NOTEBOOK_DIR = Path().resolve()

BASE_DIR   = NOTEBOOK_DIR.parent
DATA_DIR   = BASE_DIR / "data"
OUTPUT_DIR = BASE_DIR / "output"
REPORT_DIR = BASE_DIR / "report"

DATA_DIR.mkdir(parents=True, exist_ok=True)
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
REPORT_DIR.mkdir(parents=True, exist_ok=True)

print(f"BASE_DIR   : {BASE_DIR}")
print(f"DATA_DIR   : {DATA_DIR}")
print(f"OUTPUT_DIR : {OUTPUT_DIR}")
print(f"REPORT_DIR : {REPORT_DIR}")
```

## 2. Generate Data
Create synthetic training data and save to `data/`.

```python
import csv
import random
from pathlib import Path

SCRIPT_DIR = Path(__file__).resolve().parent
DATA_DIR = SCRIPT_DIR.parent / "data"
DATA_DIR.mkdir(parents=True, exist_ok=True)

random.seed(42)

def normal_noise(mu=0, sigma=1):
    """Box-Muller transform for normal distribution."""
    import math
    u1 = random.random()
    u2 = random.random()
    z = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    return mu + sigma * z

rows = []
for day_idx in range(365):
    day_of_week = day_idx % 7          # 0=Mon, 6=Sun
    is_weekend = 1 if day_of_week >= 5 else 0
    event_count = random.randint(2, 8)
    vendor_count = random.randint(20, 80)
    revenue = (12000
               + 8000 * is_weekend
               + 200 * vendor_count
               + 300 * event_count
               + normal_noise(0, 2500))
    rows.append({
        "day_of_week": day_of_week,
        "is_weekend": is_weekend,
        "event_count": event_count,
        "vendor_count": vendor_count,
        "revenue": round(revenue, 2)
    })

out_path = DATA_DIR / "revenue_data.csv"
with open(out_path, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["day_of_week", "is_weekend", "event_count", "vendor_count", "revenue"])
    writer.writeheader()
    writer.writerows(rows)

print(f"Saved {len(rows)} rows to {out_path}")
```

## 3. Train
Fit the model and save to `output/`.

```python
import csv
import pickle
import random
from pathlib import Path

SCRIPT_DIR = Path(__file__).resolve().parent
DATA_DIR = SCRIPT_DIR.parent / "data"
OUTPUT_DIR = SCRIPT_DIR.parent / "output"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# ---------- Pure Python matrix operations ----------
def _T(A):
    return [[A[j][i] for j in range(len(A))] for i in range(len(A[0]))]

def _mul(A, B):
    m, n, p = len(A), len(A[0]), len(B[0])
    return [[sum(A[i][k] * B[k][j] for k in range(n)) for j in range(p)] for i in range(m)]

def _inv(A):
    n = len(A)
    M = [A[i][:] + [1.0 if i == j else 0.0 for j in range(n)] for i in range(n)]
    for col in range(n):
        pivot = max(range(col, n), key=lambda r: abs(M[r][col]))
        M[col], M[pivot] = M[pivot], M[col]
        M[col] = [x / M[col][col] for x in M[col]]
        for row in range(n):
            if row != col:
                f = M[row][col]
                M[row] = [M[row][k] - f * M[col][k] for k in range(2 * n)]
    return [M[i][n:] for i in range(n)]

def fit_lr(X, y):
    Xb = [[1] + row for row in X]
    Xt = _T(Xb)
    theta = _mul(_inv(_mul(Xt, Xb)), _mul(Xt, [[v] for v in y]))
    return [t[0] for t in theta]

def predict_lr(theta, X):
    return [sum(t * x for t, x in zip(theta, [1] + row)) for row in X]

# ---------- Load data ----------
rows = []
with open(DATA_DIR / "revenue_data.csv", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        rows.append(row)

X_all = [[int(r["day_of_week"]), int(r["is_weekend"]), int(r["event_count"]), int(r["vendor_count"])] for r in rows]
y_all = [float(r["revenue"]) for r in rows]

# 80/20 train-test split (no shuffle to keep reproducibility)
split = int(0.8 * len(X_all))
X_train, X_test = X_all[:split], X_all[split:]
y_train, y_test = y_all[:split], y_all[split:]

theta = fit_lr(X_train, y_train)

model = {
    "theta": theta,
    "X_test": X_test,
    "y_test": y_test,
    "feature_names": ["intercept", "day_of_week", "is_weekend", "event_count", "vendor_count"]
}

out_path = OUTPUT_DIR / "revenue_forecast_model.pkl"
with open(out_path, "wb") as f:
    pickle.dump(model, f)

print(f"Model saved to {out_path}")
print(f"Theta: {[round(t, 2) for t in theta]}")
```

## 4. Evaluate
Compute metrics on the test set and save to `report/`.

```python
import pickle
import json
import math
from pathlib import Path

SCRIPT_DIR = Path(__file__).resolve().parent
OUTPUT_DIR = SCRIPT_DIR.parent / "output"
REPORT_DIR = SCRIPT_DIR.parent / "report"
REPORT_DIR.mkdir(parents=True, exist_ok=True)

def predict_lr(theta, X):
    return [sum(t * x for t, x in zip(theta, [1] + row)) for row in X]

def r2(y_true, y_pred):
    mean_y = sum(y_true) / len(y_true)
    ss_res = sum((t - p) ** 2 for t, p in zip(y_true, y_pred))
    ss_tot = sum((t - mean_y) ** 2 for t in y_true)
    return 1 - ss_res / ss_tot if ss_tot > 0 else 0.0

def mae(y_true, y_pred):
    return sum(abs(t - p) for t, p in zip(y_true, y_pred)) / len(y_true)

def rmse(y_true, y_pred):
    return math.sqrt(sum((t - p) ** 2 for t, p in zip(y_true, y_pred)) / len(y_true))

with open(OUTPUT_DIR / "revenue_forecast_model.pkl", "rb") as f:
    model = pickle.load(f)

theta = model["theta"]
X_test = model["X_test"]
y_test = model["y_test"]

y_pred = predict_lr(theta, X_test)

r2_score = r2(y_test, y_pred)
mae_score = mae(y_test, y_pred)
rmse_score = rmse(y_test, y_pred)

report = {
    "model": "Multiple Linear Regression",
    "task": "Admin daily revenue forecast",
    "train_samples": 292,
    "test_samples": len(y_test),
    "r2": round(r2_score, 4),
    "mae": round(mae_score, 2),
    "rmse": round(rmse_score, 2)
}

out_path = REPORT_DIR / "evaluation_report.json"
with open(out_path, "w") as f:
    json.dump(report, f, indent=2)

print(f"Report saved to {out_path}")
print(f"R²={r2_score:.4f}  MAE={mae_score:.2f}  RMSE={rmse_score:.2f}")
```

## 5. Test
Run inference on sample inputs.

```python
import pickle
import json
from pathlib import Path

SCRIPT_DIR = Path(__file__).resolve().parent
OUTPUT_DIR = SCRIPT_DIR.parent / "output"
REPORT_DIR = SCRIPT_DIR.parent / "report"

def predict_lr(theta, X):
    return [sum(t * x for t, x in zip(theta, [1] + row)) for row in X]

def r2(y_true, y_pred):
    mean_y = sum(y_true) / len(y_true)
    ss_res = sum((t - p) ** 2 for t, p in zip(y_true, y_pred))
    ss_tot = sum((t - mean_y) ** 2 for t in y_true)
    return 1 - ss_res / ss_tot if ss_tot > 0 else 0.0

with open(OUTPUT_DIR / "revenue_forecast_model.pkl", "rb") as f:
    model = pickle.load(f)

theta = model["theta"]
X_test = model["X_test"]
y_test = model["y_test"]

# Sample inputs: Mon-Sun, 5 events, 50 vendors
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
sample_X = []
for dow in range(7):
    is_wknd = 1 if dow >= 5 else 0
    sample_X.append([dow, is_wknd, 5, 50])

preds = predict_lr(theta, sample_X)

print("\n=== Admin Revenue Forecast: Mon-Sun Predictions (5 events, 50 vendors) ===")
print(f"{'Day':<5} {'Predicted Revenue':>20}")
for day, pred in zip(days, preds):
    print(f"{day:<5} ${pred:>19,.2f}")

y_pred_test = predict_lr(theta, X_test)
r2_val = r2(y_test, y_pred_test)
print(f"\nActual R² on test set: {r2_val:.4f}")

# Load report
with open(REPORT_DIR / "evaluation_report.json") as f:
    report = json.load(f)
print(f"From report — R²: {report['r2']}, MAE: {report['mae']}, RMSE: {report['rmse']}")
```
